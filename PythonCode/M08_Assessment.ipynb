```python
#Characteristics of Businesses
import pandas as pd
import matplotlib.pyplot as plt
import requests
import seaborn as sns
```

```python
#Question1
#Number of Hispanic and nonEmployees

cbq1 = cb.rename(columns={"ETH_GROUP_LABEL":"ethnicgrouplab","EMP":"numemployees","SEX_LABEL":"sexlabel","RACE_GROUP_LABEL":"racelabel"})

#Question2 
#Number of Employees by Employer firms
```

```python
cbq1['numemployees']=cbq1['numemployees'].astype(float)
```

```python
cbq1.dtypes
```

```python
cbq1 = cbq1[['sexlabel','ethnicgrouplab','racelabel','numemployees']]
```

```python
cbq1
```

```python
cbh = cbq1[cbq1.ethnicgrouplab!='Equally Hispanic/non-Hispanic']
cbh = cbh[cbq1.ethnicgrouplab!='Unclassifiable']
cbh = cbh[cbq1.ethnicgrouplab!='Classifiable']
#cbh = cbh[cbq1.ethnicgrouplab!='Total']
cbh = cbh[cbq1.sexlabel!='Total']
cbh = cbh[cbq1.sexlabel!='Equally male/female']
cbh
```

```python
cbh=cbh[['ethnicgrouplab','numemployees']]
ree=cbh.groupby(['ethnicgrouplab']).count()
```

```python
cbq1['ethnicgrouplab'].unique()
```

```python
plt.figure(figsize=(12,14))
sns.set_style("darkgrid")
ax = ree.plot(title = 'Number of Employees that are Hispanic and non-Hispanic',kind='bar')
plt.xlabel('Ethnic Group')
plt.ylabel('Count of Employees')
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0], label_type='edge', fmt='%d');
```

```python
#Question 3
#Is there a relationship between the number of employees and the number of owners
```

```python
#Load characteristics of business
url2 = 'https://api.census.gov/data/2018/abscbo?get=GEO_ID,NAME,NAICS2017,NAICS2017_LABEL,OWNER_SEX,OWNER_SEX_LABEL,OWNER_ETH,OWNER_ETH_LABEL,OWNER_RACE,OWNER_RACE_LABEL,OWNER_VET,OWNER_VET_LABEL,QDESC,QDESC_LABEL,OWNCHAR,OWNCHAR_LABEL,YEAR,OWNPDEMP,OWNPDEMP_F,OWNPDEMP_PCT,OWNPDEMP_PCT_F,OWNPDEMP_S,OWNPDEMP_S_F,OWNPDEMP_PCT_S,OWNPDEMP_PCT_S_F&for=us:*&QDESC_LABEL=YRACQBUS&key=b3ba78a4a01bf0ea2af20c82bf42f0e425341ee5'
response2 = requests.get(url2)
data2 = response2.json()
cbo = pd.DataFrame(data2,columns=data2.pop(0))
```

```python
cbo
```

```python
merged = cb.merge(cbo,how='inner',on='NAICS2017')
```

```python
merged
```

```python
merged.columns
```

```python
merged['RACE_GROUP_LABEL'].unique()
```

```python
merged['OWNER_RACE_LABEL']
```

```python
merged['EMP']=merged['EMP'].astype(float)
merged['OWNPDEMP']=merged['OWNPDEMP'].astype(float)
merged=merged.groupby(['NAICS2017_LABEL_x']).count()



sns.set_theme(style="darkgrid")
plt.figure(figsize = (20,10))
plt.title('Number of Employees vs Number of Owners', size=22)
sns.scatterplot(x='OWNPDEMP',
                y='EMP',
                data=merged,
                s=300,
                color='red',
               alpha=0.2,
               edgecolor='black')
plt.ylim([0,250000])
plt.xlim([0,250000])
plt.tick_params(axis='x', labelsize=16)
plt.tick_params(axis='y', labelsize=16)
```
